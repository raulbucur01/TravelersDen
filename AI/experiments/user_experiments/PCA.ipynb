```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt

df = pd.read_csv("../../data/processed/user_data_prepared.csv")

# Assuming your data is loaded in 'df' and interests are in the columns 'Interest1', 'Interest2', ..., 'InterestN'
# Extract the features (interest columns)
X = df.drop(['UserID', 'Gender', 'Country', 'AgeGroup'], axis=1)  # Drop non-feature columns

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA
pca = PCA(n_components=2)  # Reduce to 2 components for visualization
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame with the PCA components
pca_df = pd.DataFrame(data=X_pca, columns=['PCA1', 'PCA2'])

# Plot the PCA results
plt.figure(figsize=(8, 6))
plt.scatter(pca_df['PCA1'], pca_df['PCA2'], alpha=0.5)
plt.title("PCA of User Interests")
plt.xlabel("PCA1")
plt.ylabel("PCA2")
plt.show()

# Explained variance ratio
print("Explained variance ratio by component:", pca.explained_variance_ratio_)
```

```python
import pandas as pd
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
import matplotlib.pyplot as plt
from mpl_toolkits.mplot3d import Axes3D

# Assuming your data is loaded in 'df' and interests are in the columns 'Interest1', 'Interest2', ..., 'InterestN'
# Extract the features (interest columns)
X = df.drop(['UserID', 'Gender', 'Country', 'AgeGroup'], axis=1)  # Drop non-feature columns

# Standardize the data
scaler = StandardScaler()
X_scaled = scaler.fit_transform(X)

# Apply PCA with 3 components
pca = PCA(n_components=3)
X_pca = pca.fit_transform(X_scaled)

# Create a DataFrame with the PCA components
pca_df = pd.DataFrame(data=X_pca, columns=['PCA1', 'PCA2', 'PCA3'])

# Plot the PCA results in 3D
fig = plt.figure(figsize=(10, 7))
ax = fig.add_subplot(111, projection='3d')
ax.scatter(pca_df['PCA1'], pca_df['PCA2'], pca_df['PCA3'], alpha=0.5)
ax.set_title("PCA of User Interests (3 Components)")
ax.set_xlabel("PCA1")
ax.set_ylabel("PCA2")
ax.set_zlabel("PCA3")

plt.show()

# Explained variance ratio
print("Explained variance ratio by component:", pca.explained_variance_ratio_)
```
